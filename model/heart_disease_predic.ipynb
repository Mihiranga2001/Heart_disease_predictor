```python
data = pd.read_csv('heart_disease.csv')
data.head()
```

```python
# drop null raws
data = data.dropna()
```

```python
data.shape
```

```python
data.info()
```

```python
data['Heart Disease Status'] = data['Heart Disease Status'].map({'Yes': 1, 'No': 0})
```

```python
numeric_data = data.select_dtypes(include=['int64', 'float64'])
numeric_data.corr()['Heart Disease Status']
```

```python
data.tail()
```

```python
data.tail()
```

```python
data['Gender'] = data['Gender'].replace({'Male': 1, 'Female': 0})
```

```python
data.head()
```

```python
data['Smoking'] = data['Smoking'].replace({'Yes': 1, 'No': 0})
data['Family Heart Disease'] = data['Family Heart Disease'].replace({'Yes': 1, 'No': 0})
data['Diabetes'] = data['Diabetes'].replace({'Yes': 1, 'No': 0})
data['High Blood Pressure'] = data['High Blood Pressure'].replace({'Yes': 1, 'No': 0})
data['Low HDL Cholesterol'] = data['Low HDL Cholesterol'].replace({'Yes': 1, 'No': 0})
data['High LDL Cholesterol'] = data['High LDL Cholesterol'].replace({'Yes': 1, 'No': 0})
data['Exercise Habits'] = data['Exercise Habits'].replace({'High': 3, 'Medium':2, 'Low':1, 'None': 0})
data['Alcohol Consumption'] = data['Alcohol Consumption'].replace({'High': 3, 'Medium':2, 'Low':1, 'None': 0})
data['Stress Level'] = data['Stress Level'].replace({'High': 3, 'Medium':2, 'Low':1, 'None': 0})
data['Sugar Consumption'] = data['Sugar Consumption'].replace({'High': 3, 'Medium':2, 'Low':1, 'None': 0})
```

```python
data.head()
```

```python
numeric_data = data.select_dtypes(include=['int64', 'float64'])
numeric_data.corr()['Heart Disease Status']
```

```python
data.shape
```

```python
X = data.drop('Heart Disease Status', axis=1)
y = data['Heart Disease Status']
```

```python
%pip install imbalanced-learn
#SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote,y_smote = smote.fit_resample(X,y)
y_smote.value_counts()
```

```python
%pip install sklearn
X = x_smote
y = y_smote
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc = model.score(X_test,y_test)
    print(str(model)+'-->'+str(acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model_acc(rf)

from sklearn.svm import SVC
svm = SVC()
model_acc(svm)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
model_acc(knn)

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model_acc(nb)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
model_acc(lr)

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
model_acc(dt)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_obj = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid_fit = grid_obj.fit(X_train,y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test,y_test)
```

```python
import pickle

with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
